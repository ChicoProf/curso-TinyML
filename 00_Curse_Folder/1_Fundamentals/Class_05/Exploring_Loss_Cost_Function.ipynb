Creating a function

```python
def calc_plot_error_ORIGINAL (X, Y, w, b):
    Y_hat = []
    for Xi in X:
        Yi = (w*Xi)+b
        Y_hat.append(Yi)

    total_squared_error = 0
    n = len(Y)
    for i in range(0, n):
        squared_error = (Y[i] - Y_hat[i]) ** 2
        total_squared_error += squared_error

    MSE = (1/n)*(total_squared_error)
    print("MSE is: {:.2f}".format(MSE))
    grafico.plot(X, Y, color='red', marker='o', label='Y')
    grafico.plot(X, Y_hat, color='blue', marker='o', label='predictions')
    grafico.axis([-1,5,-5,15])
    grafico.title('Predictions')
    grafico.xlabel('X (input data)')
    grafico.ylabel('Y (output)')
    #plt.legend(loc='upper left')
    grafico.grid()

    return MSE

def calc_plot_error(X,Y,w,b):
  Y_hat=[w*Xi+b for Xi in X]
  erro=[(Y[i]-Y_hat[i])**2 for i in range(len(Y))]
  total_squared_error=sum(erro)
  MSE=sum(erro)/n
  print(f"O Erro Médio Quadrático é {MSE:.2f}")
  grafico.plot(X,Y,color='green', marker='o',label='Y (valores reais)')
  grafico.plot(X, Y_hat,color='red', marker='x', label=f"estimativa para w={w} e b={b}")
  grafico.axis([min(X)-1,max(X)+1,min(Y)-1,max(Y)+1])
  grafico.title("Previsões")
  grafico.xlabel("X (dados de entrada)")
  grafico.ylabel("Dados de saída")
  grafico.legend(loc='upper left')
  grafico.grid()

  return round(MSE,2)
```
